```python
import pandas as pd
data = {
    'Name': ['Pavan','Kapil','Lalit','Ishan','Om'],
    'Age': [25, None, 44, 23, None],
    'Salary': [50000, 60000, 70000, None , None]
}
df = pd.DataFrame(data)
print("Original DataFrame")
print(df)

print(df.isnull().sum()) #shows the missing places
df_drop= df.dropna() #drops any column/row that has an empty space
print(df_drop)

df['Age'].fillna(df['Age'].mean(), inplace=True) #fills the empty spaces with mean of others.
df['Salary'].fillna(df['Salary'].mean(), inplace=True) #same as above.
print(df)

print(df.isnull().mean() * 100) #tells the total percentage of data that is missing.
```
